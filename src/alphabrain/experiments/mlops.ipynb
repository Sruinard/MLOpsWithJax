```python
from azure.ai.ml.entities import Model
from azure.ai.ml.constants import ModelType
repo = azure_ml_repo.AzureMLRepo()
env_name = "jaxserving"
version = "1"
serving_env = repo.ml_client.environments.get(name=env_name, version=version)

model_artifact_name = "alphabrain"
model_version = "6"
model_name = "brain"
#atastores/workspaceblobstore/paths/azureml/8b611fb8-afe0-4464-b6a7-eba88541e8b6/serving_model_dir/brain/
# azureml://subscriptions/0abb6ec5-9030-4b3f-af04-09183c688576/resourcegroups/csu-nl-intelligence/workspaces/mlpatterns/datastores/workspaceblobstore/paths/azureml/8b611fb8-afe0-4464-b6a7-eba88541e8b6/serving_model_dir/brain/20221101080400/
serving_model_dir_input = "azureml://subscriptions/0abb6ec5-9030-4b3f-af04-09183c688576/resourcegroups/csu-nl-innovative-ml-apps/workspaces/alphabrain-platform-auofwmqbg5jr2/datastores/workspaceblobstore/paths/azureml/d94519dc-ca4e-4484-91b4-7c3e817a0355/serving_model_dir/"
path_to_ml_model_on_storage = f"{serving_model_dir_input}{model_name}"
# datastore_relative_path = serving_model_dir_input.split("datastores/")[-1]
# path_to_ml_model_on_storage = f"azureml://datastores/{datastore_relative_path}/{model_name}"
model = Model(
    path=path_to_ml_model_on_storage,
    name=model_artifact_name,
    version=model_version,
    type="custom_model",
    description="Model created from pipeline run."
)
repo.ml_client.models.create_or_update(model)
```

```python

from azure.ai.ml.entities import ManagedOnlineDeployment
from alphabrain.mlops.trainer import azure_ml_repo
from alphabrain.config import PipelineConfig

model = repo.ml_client.models.get(name=model_artifact_name, version=model_version)

deployment_name = 'brain-deployment-green-with-wait'
blue_deployment = ManagedOnlineDeployment(
    name=deployment_name,
    endpoint_name=PipelineConfig.deployment_config.online_endpoint_name,
    model=model,
    environment=serving_env,
    environment_variables={
        "MODEL_BASE_PATH": f"/var/azureml-app/azureml-models/{model_artifact_name}/{model_version}",
        "MODEL_NAME": model_name,
    },
    instance_type="Standard_DS2_v2",
    instance_count=1,
    
)

endpoint = repo.ml_client.online_endpoints.get(name=PipelineConfig.deployment_config.online_endpoint_name)
endpoint.traffic = {deployment_name: 100}
```

```python

repo.ml_client.begin_create_or_update(blue_deployment).result()
repo.ml_client.begin_create_or_update(endpoint)
```
